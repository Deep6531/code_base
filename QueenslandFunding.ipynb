```python
import pandas as pd

#  file path
file1 = '/content/advance-queensland-funding-recipients.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file1, encoding = "ISO-8859-1")

df
```

```python
df.head()
```

```python
df.set_index('Recipient Name', inplace=True)
df

```

```python
df.dtypes
```

```python
# Assuming your DataFrame is named df
# Convert the "Actual Contractual Commitment ($)" column to numeric if it's not already
df['Actual Contractual Commitment ($)'] = pd.to_numeric(df['Actual Contractual Commitment ($)'], errors='coerce')

# Group the DataFrame by the "Program" column and calculate the sum of the funding amounts
total_funding_per_program = df.groupby('Program')['Actual Contractual Commitment ($)'].sum()

# Print the result
print(total_funding_per_program)
```

```python
# Assuming 'total_funding_per_program' is the Series containing the grouped data
# Convert the Series to a DataFrame
grouped_data_df = total_funding_per_program.reset_index()
grouped_data_df.columns = ['Program', 'Total Funding']

# Save the DataFrame as a CSV file
grouped_data_df.to_csv('total_funding_per_program.csv', index=False)

# Print a message to confirm that the file has been saved
print("Data saved successfully.")
```

```python
import matplotlib.pyplot as plt

# Plot the grouped data as a bar plot
plt.figure(figsize=(12, 8))
plt.bar(grouped_data_df['Program'], grouped_data_df['Total Funding'])

# Add labels and title
plt.xlabel('Program')
plt.ylabel('Total Funding ($)')
plt.title('Total Funding Amount for Each Program')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.tight_layout()
plt.show()
```
